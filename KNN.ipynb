```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# read date
info = pd.read_csv('.\log\\100.csv')
```

```python
# L0 distance
fig = sns.jointplot(x='k', y='accu', data=info[:100], kind='reg')
fig.fig.set_size_inches(3,2)
plt.show()
```

```python
# L1 distance
fig = sns.jointplot(x='k', y='accu', data=info[100:200], kind='reg')
fig.fig.set_size_inches(3,2)
plt.show()
```

```python
# L2 distance
fig = sns.jointplot(x='k', y='accu', data=info[200:300], kind='reg')
fig.fig.set_size_inches(3,2)
plt.show()
```

```python
fig = sns.jointplot(x='k', y='accu', data=info, 
                    kind='kde',hue='distance', 
                    shade=True,joint_kws=dict(alpha=0.6))
fig.fig.set_size_inches(6,5)
plt.show()
```

```python
# heat map
# annotation 标注
sns.heatmap(info.corr(), annot=True, cmap='coolwarm')
```
